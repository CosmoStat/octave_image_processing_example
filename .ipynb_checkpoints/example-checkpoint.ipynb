# Exercice: Restauration d'Image avec Critère de Qualité

Les images peuvent être corrompues par du bruit qui peut s'être introduit lors de la génération, de la transmission ou du stockage de celles-ci. Dans cette partie nous nous intéressons à la restauration des images bruitées, en utilisant le Rapport Signal à Bruit.

## Critère de Qualité: Rapport Signal à Bruit (RSB)

Le RSB (Signal to Noise Ratio, SNR en anglais) est un critère de qualité souvent utilisé pour restaurer les images surtout que le nombre de celles-ci ne permet pas de vérifier visuellement chacune d'elles. L'idée est de ce critère est de quantifier la quantité de signal utile et la quantité de bruit dans l'image puis de faire le rapport. Pour cela, considérons une image bruitée $Y$ contenant le signal utile (càd l'image sans bruit) $X$ et le bruit $N$, alors le RSB de $Y$ s'écrit:
\begin{equation}
\text{RSB}(Y)=\frac{\|X\|_2}{\|N\|_2} \quad,
\end{equation}
avec $\|\cdot\|_2$ la norme $\ell_2$, par exemple $\|X\|_2=\sqrt{\sum_{i,j}{X[i,j]^2}}$.

## Partie 1: Détermination de la Taille Idéale du Filtre.

Dans cet exercice, nous allons débruitée une image en changeant la taille.

**Question 1:** Chargez la variable `im_b` du ficher `canards_gauss.mat` et afficher l'image.
> Pour charger et renommer une variable à partir d'un fichier .mat, utilisez la commande `ma_var = load("le_fichier.mat").la_var`.


```octave
%chargement de l'image
im_b=load("canards_gauss.mat").im_b;

%affichage
figure(1)
%chargement de la carte de couleur
colormap("gray")
imagesc(im_b);
axis('off');
title('Image avec bruit') ;
```

**Question 2:** Sachant que le bruit est additif, càd que $image$ $bruitée = image$ $vraie+bruit$. Extrayez le bruit et affichez-le. Il s'agit d'un bruit blanc gaussien. Observez-vous une structure en particulier?

```octave
%chargement de l'image vraie
im_v = load("canards.mat").im;
bruit = im_b-im_v;

%affichage
figure(2)
colormap("gray")
imagesc(bruit);
axis('off');
title('Bruit') ;
```

**Question 3** Calculez le RSB de l'image bruitée et donnez sa valeur en dB.
> Astuce: Pour sommer tous les éléments de la matrice A, écrire `sum(A(:))`, tel que A(:) est un vecteur contenant tous les éléments de A.

```octave
RSB = sqrt(sum((im_v.^2)(:))/sum((bruit.^2)(:)))
RSB_db = 10*log10(RSB)
```

Maintenant nous souhaitons réduire le bruit nous allons utiliser un filtre moyenneur. Ce filtre remplace la valeur de chaque pixel dans l'image par celle de la moyenne des pixels voisins à ce pixels contenus dans un carré de taille $n \times n$ et centré sur le pixel en question. Un filtre moyenneur a donc la forme suivante:
$$M_n = \frac{1}{n}\cdot\begin{bmatrix} \frac{1}{n} & \dots & \frac{1}{n} \\ \vdots & \ddots & \vdots \\ \frac{1}{n} & \dots & \frac{1}{n} \end{bmatrix}\quad.$$
Pour appliquer ce filtre sur une image, $I$, il suffit d'effectuer leur produit de convolution, notons le $G$, de manière analogue au cas 1-D:
$$G = I \ast M_n \quad. $$

**Question 4** Générez un filtre moyenneur de taille $n=20$ avec la commande `M20=moyenneur(20);`. Convoluez celui-ci avec l'image bruitée. Affichez le résultat et calculer son RSB en dB. Comparer ce RSB à celui de l'image bruitée.
> Pour convoluer une image A par un filtre F, utilisez la fonction `conv2` comme suit `res = conv2(A,F,"same")`. Attention, l'ordre à bien mettre l'image en premier, ensuite le filtre!

```octave
%filtrage
M20 = moyenneur(20);
res20 = conv2(im_b,M20,"same");
%calcul du bruit restant
bruit20 = res20-im_v;

%affichage
figure(3)
colormap("gray")
imagesc(res20);
axis('off');
title('Image filtrée (taille 20)') ;
%calcul du RSB
RSB20 = sqrt(sum((im_v.^2)(:))/sum((bruit20.^2)(:)))
RSB20_db = 10*log10(RSB20)
```

**Question 5** Refaites la question précédente pour les tailles $n=5$ et $n=50$. Quelle est l'impact de la taille du filtre sur le résultat, qualitativement et quantitativement? Quelle est la taille idéale du filtre parmi ces 3 valeurs?

```octave
%Pour n=5
%filtrage
M5 = moyenneur(5);
res5 = conv2(im_b,M5,"same");
%calcul du bruit restant
bruit5 = res5-im_v;

%affichage
figure(4)
colormap("gray")
imagesc(res5);
axis('off');
title('Image filtrée (taille 5)') ;
%calcul du RSB
RSB5 = sqrt(sum((im_v.^2)(:))/sum((bruit5.^2)(:)))
RSB5_db = 10*log10(RSB5)

%Pour n=50
%filtrage
M50 = moyenneur(50);
res50 = conv2(im_b,M50,"same");
%calcul du bruit restant
bruit50 = res50-im_v;

%affichage
figure(5)
colormap("gray")
imagesc(res50);
axis('off');
title('Image filtrée (taille 50)') ;
%calcul du RSB
RSB50 = sqrt(sum((im_v.^2)(:))/sum((bruit50.^2)(:)))
RSB50_db = 10*log10(RSB50)
```


## Partie 2: Choix du Filtre Suivant la Nature du Bruit

Dans cet exercice, nous allons comparer 3 filtres: moyenneur, min et max. Lorsque l'un de ces filtres est appliqué sur une image, la valeur de chaque pixel de l'image est remplacée par la sortie de la fonction portant le nom du filtre. Celle-ci est alors appliquée sur les pixels voisins contenus dans un carré de taille $n \times n$ et centré sur le pixel considéré.  Avec ceci, nous allons tenter d'éliminer deux images contaminées chacune par un bruit de nature différente.

### Bruit Sel

Lorsqu'on fait tomber du sel sur une image, nous voyons apparaître des points blancs dans l'image. Ainsi les images contaminées par un bruit sel sont des images contenant des pixels blancs. Ce bruit peut provenir par exemple, de capteurs dans les caméras saturent (lorsqu'ils reçoivent une intensité très élevées). Dans la suite, nous allons essayer différentes méthodes pour taiter ce bruit et le bruit blanc gaussien.

**Question 6** Chargez la variable `im_sel` du ficher `canards_sel.mat` et afficher l'image.

```octave
%chargement de l'image
im_sel=load("canards_sel.mat").im_sel;

%affichage
figure(6)
colormap("gray")
imagesc(im_sel);
axis('off');
title('Image avec un bruit sel') ;
```

**Question 7** Sachant que le bruit sel est additif, càd que $image$ $bruitée = image$ $vraie+bruit$. Extrayez le bruit et affichez-le. Calulez le RSB de l'image bruitée.

```octave
sel = im_sel-im_v;

%affichage
figure(7)
colormap("gray")
imagesc(sel);
axis('off');
title('Bruit Sel');

%calcul du RSB
RSBs = sqrt(sum((im_v.^2)(:))/sum((sel.^2)(:)))
RSBs_db = 10*log10(RSBs)
```

**Question 8** Appliquez le filtre moyenneur de taille $n=15$ sur les images `im_b` (avec bruit gaussien) et l'image `im_sel` (avec bruit sel). Affichez les images résultantes et calculez le RSB de chacune d'elles. Que remarquez-vous, qualitativement et quantitativement?

```octave
%generation du filtre
M = moyenneur(15);

%Pour bruit gaussien
%filtrage
im_bM = conv2(im_b,M,"same");
%calcul du bruit restant
residu_b = im_bM-im_v;

%affichage
figure(8)
colormap("gray")
imagesc(im_bM);
axis('off');
title('Image filtrée (bruit gaussien)') ;
%calcul du RSB
RSBbM = sqrt(sum((im_v.^2)(:))/sum((residu_b.^2)(:)))
RSBbM_db = 10*log10(RSBbM)

%Pour bruit sel
%filtrage
im_sM = conv2(im_sel,M,"same");
%calcul du bruit restant
residu_sM = im_sM-im_v;

%affichage
figure(9)
colormap("gray")
imagesc(im_sM);
axis('off');
title('Image filtrée (bruit sel)') ;
%calcul du RSB
RSBsM = sqrt(sum((im_v.^2)(:))/sum((residu_sM.^2)(:)))
RSBsM_db = 10*log10(RSBsM)
```
